```python
import numpy as np
import cv2
```

```python
model=hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

# 1.preprocess Image and Load

```python
def load_image(img_path):
    img=tf.io.read_file(img_path)
    img=tf.image.decode_image(img,channels=3)
    img=tf.image.convert_image_dtype(img,tf.float32)
    img=img[tf.newaxis,:]
    return img
```

```python
content_img=load_image('cloth.jpeg')
style_image=load_image('360_F_175360692_ADvH306d4gL4sAsRcU3mxbb1z4umWb4W.jpg')
```

# 2. visualize output

```python
styled_img=model(tf.constant(content_img), tf.constant(style_image))[0]
```
